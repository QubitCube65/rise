````python
# of course you can show figures
%pip install numpy matplotlib
import numpy as np
import matplotlib.pyplot as plt

def polynom(x):
    return 2 * x**2 - 20 * x + 2

X = np.linspace(-10, 10)
Y = polynom(X)
plt.plot(X, Y);
````

````python
# and everything works as usual

# an animation to illustrate 
# translation by variable change
import ipywidgets as widgets
from ipywidgets import interact, FloatSlider
import matplotlib.pyplot as plt



def parabolic(offset):
    X = np.linspace(-10, 10)
    Y = polynom(X-offset)
    # use same y scale for all offsets
    
    plt.gca().set_ylim([-100, 500])
    plt.plot(X, Y);
    
interact(parabolic, 
         offset=FloatSlider(min=-10., max=10.,
                           step=0.25));
````

# Publishing on binder

In order for a binder-hosted notebook to start in slideshow mode, you need to have the following tag set in the notebook metadata:

```javascript        
    ...
    "rise": {
        "autolaunch": true
    }
    ...
```

You can edit the notebook metadata in the the `Proberty Inspector` menu,  under Advance Tools  submenu `Notebook metadata`.

Note finally that the `rise` key in this json file used to be named `livereveal`. The latter is still honored, but the former takes precedence, and it is recommended to use only `rise` from now on.

# Decorating all slide
RISE offers two levels for inserting a static background. You can either define overlay, in which case you take full control,

or you can define header, footer and backimage. So if you define overlay, the 3 latter options will be ignored.

## overlay
It is possible to add the config option overlay to build a constant background. 
It is wrapped in a<div>, so it can be text or html. 
In this case, the user is entirely responsible for styling. For example:

```
{
 ...
 "rise": {
     "overlay": "<div class='myheader'><h2>my company</h2></div><div class='myfooter'><h2>the date</h2></div>"
 }
}
```


## header, footer and backimage
As a more limited, but often more convenient alternative, you can define any of the following 3 settings.

In this case, minimal styling is applied (floor and ceiling), but user is still responsible for cosmetic styling:
``` 
 ...
"rise": {
        "backimage": "examples/mybackimage.png",
        "header": "<div style='position: fixed; top: 10px; left: 20px; font-size: 18px; color: #333; z-index: 1000;'>ðŸ“Š My Presentation</div>",
        "footer": "<div style='position: fixed; bottom: 10px; left: 20px; font-size: 14px; color: #666; z-index: 1000;'>&copy; Moderator</div><div style='position: fixed; bottom: 10px; right: 20px; font-size: 14px; color: #666; z-index: 1000;'>Page <span class='slide-number'></span></div>"
    }

```

## Background Image:

### Background Image for the Entire Presentation (via CSS):
To apply a background image to all slides in a RISE presentation, add the following rule to your CSS file:
```
{
body.rise-enabled .reveal .slides section {
  background-image: url("mybackimage.png");
  background-size: cover;
  background-position: center;
  background-repeat: no-repeat;
}

```
Import the CSS file in a Markdown cell at the top of your notebook:

```
<style>
@import url("example.css");
</style>

```
Once loaded, the background image will be applied automatically when entering RISE presentation mode.


### Slide-Specific Background Image:
To apply a background image to a single slide only, use a slide-specific CSS class.

Define the Background in CSS:
Refer to README.css for additional customization options.

Afterwards use the Slide in Markdown cell:

